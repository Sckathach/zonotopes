```python
from zonotope.utils import create_zonotope
from zonotope.plot import plot_zonotope_2d

z = create_zonotope(
    center_values=[4, 3],
    infinity_terms=[[-1, 2], [1, 1]],
    special_terms=[[1, 0], [0, 1]],
)
plot_zonotope_2d(
    z, plot_without_special_terms=True, title=" ", arrow_size=0, savefile="test.pdf"
)
```
