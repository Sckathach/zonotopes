```python
import torch as t
import einx

a = t.arange(16).view(4, 4).float()
a[:, 1] = 3
b = t.nn.functional.softmax(a, dim=-1)
a, b
```

```python
a_ = einx.rearrange("a b -> a b c", a, c=a.shape[-1])
a_diff = a_ - einx.rearrange("a b c -> a c b", a_)
a_diff.shape
```

```python
a_diff_exp = a_diff.exp()
1 / a_diff_exp.sum(-2)
```
